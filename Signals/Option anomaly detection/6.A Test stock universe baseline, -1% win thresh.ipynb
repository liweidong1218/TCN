```python
import pickle 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from anomaly_detection_helper import train_test_split
plt.style.use('seaborn')
```

```python
import os
import sys

project_dir = os.path.realpath(os.path.join(os.getcwd(), '..','..','..','..'))
tickerdata_dir = os.path.join(project_dir, 'ORATS','byticker')
symboluniv_data_file = [os.path.join(tickerdata_dir, f) for f in os.listdir(tickerdata_dir)]
symboluniv_data_file
```

```python
with open('symbol.pkl','rb') as f:
    symbol_arr = pickle.load(f)
with open('test_performance.pkl','rb') as f:
    test_perform = pickle.load(f)
```

```python
a2_summary = pd.concat(test_perform['A2'])
a2_summary['ticker'] = symbol_arr
a2_summary.dropna(how='any',inplace=True)
a2_summary.reset_index(inplace=True, drop=True)

b2_summary = pd.concat(test_perform['B2'])
b2_summary['ticker'] = symbol_arr
b2_summary.dropna(how='any',inplace=True)
b2_summary.reset_index(inplace=True, drop=True)

b3_summary = pd.concat(test_perform['B3'])
b3_summary['ticker'] = symbol_arr
b3_summary.dropna(how='any',inplace=True)
b3_summary.reset_index(inplace=True, drop=True)
```

```python
def winrate(stock_series, win_thresh=-0.01, lag=5):
    lag_series = stock_series.shift(lag)
    ret = np.log(stock_series / lag_series)
    win_rate = np.mean(ret>win_thresh)
    return round(win_rate,3)
```

```python
#1st data file
symbolUniv = pd.read_parquet(symboluniv_data_file[0])
```

```python
split_date = '2019-06-01'
window = 60
for symbol in symbolUniv['underlying'].unique():
    symbolData = symbolUniv[symbolUniv['underlying']==symbol]
    try:
        _, test = train_test_split(symbolData, split_date, window)
    except:
        print(symbol, 'no enough data')
        continue
    stock_series = test.groupby('quotedate')['underlying_last'].first()
    win_rate = winrate(stock_series)
    a2_summary.loc[a2_summary['ticker'] == symbol,'winrate_alldate'] = win_rate
    b2_summary.loc[b2_summary['ticker'] == symbol,'winrate_alldate'] = win_rate
    b3_summary.loc[b3_summary['ticker'] == symbol,'winrate_alldate'] = win_rate
```
